```python
# 진해경찰서 2020(상반기)

df[368-41:369-41]['popu_u20'] = df.iloc[368 - 41][2] * df.iloc[368-82][4] / df.iloc[368-82][2]
df[368-41:369-41]['popu_2030'] = df.iloc[368 - 41][2] * df.iloc[368-82][5] / df.iloc[368-82][2]
df[368-41:369-41]['popu_4050'] = df.iloc[368 - 41][2] * df.iloc[368-82][6] / df.iloc[368-82][2]
df[368-41:369-41]['popu_o60'] = df.iloc[368 - 41][2] * df.iloc[368-82][7] / df.iloc[368-82][2]

df[368-41:369-41]['for_u20'] = df.iloc[368 - 41][3] * df.iloc[368-82][8] / df.iloc[368-82][3]
df[368-41:369-41]['for_2030'] = df.iloc[368 - 41][3] * df.iloc[368-82][9] / df.iloc[368-82][3]
df[368-41:369-41]['for_4050'] = df.iloc[368 - 41][3] * df.iloc[368-82][10] / df.iloc[368-82][3]
df[368-41:369-41]['for_o60'] = df.iloc[368 - 41][3] * df.iloc[368-82][11] / df.iloc[368-82][3]
```

```python
# 진해경찰서 2020(하반기)

df[368:]['popu_u20'] = df.iloc[368 - 41][2] * df.iloc[368-82][4] / df.iloc[368-82][2]
df[368:]['popu_2030'] = df.iloc[368 - 41][2] * df.iloc[368-82][5] / df.iloc[368-82][2]
df[368:]['popu_4050'] = df.iloc[368 - 41][2] * df.iloc[368-82][6] / df.iloc[368-82][2]
df[368:]['popu_o60'] = df.iloc[368 - 41][2] * df.iloc[368-82][7] / df.iloc[368-82][2]

df[368:]['for_u20'] = df.iloc[368 - 41][3] * df.iloc[368-82][8] / df.iloc[368-82][3]
df[368:]['for_2030'] = df.iloc[368 - 41][3] * df.iloc[368-82][9] / df.iloc[368-82][3]
df[368:]['for_4050'] = df.iloc[368 - 41][3] * df.iloc[368-82][10] / df.iloc[368-82][3]
df[368:]['for_o60'] = df.iloc[368 - 41][3] * df.iloc[368-82][11] / df.iloc[368-82][3]
```

```python
df
```

```python
# 나머지경찰서 2020(상반기)
for i in range(359, 368):
    df[i-41:i-40]['popu_u20'] = df.iloc[i - 41]['popu_합계'] * df.iloc[i - 82]['popu_u20'] / df.iloc[i - 82]['popu_합계']
    df[i-41:i-40]['popu_2030'] = df.iloc[i - 41]['popu_합계'] * df.iloc[i - 82]['popu_2030'] / df.iloc[i - 82]['popu_합계']
    df[i-41:i-40]['popu_4050'] = df.iloc[i - 41]['popu_합계'] * df.iloc[i - 82]['popu_4050'] / df.iloc[i - 82]['popu_합계']
    df[i-41:i-40]['popu_o60'] = df.iloc[i - 41]['popu_합계'] * df.iloc[i - 82]['popu_o60'] / df.iloc[i - 82]['popu_합계']

    df[i-41:i-40]['for_u20'] = df.iloc[i - 41]['for_합계'] * df.iloc[i - 82]['for_u20'] / df.iloc[i - 82]['for_합계']
    df[i-41:i-40]['for_2030'] = df.iloc[i - 41]['for_합계'] * df.iloc[i - 82]['for_2030'] / df.iloc[i - 82]['for_합계']
    df[i-41:i-40]['for_4050'] = df.iloc[i - 41]['for_합계'] * df.iloc[i - 82]['for_4050'] / df.iloc[i - 82]['for_합계']
    df[i-41:i-40]['for_o60'] = df.iloc[i - 41]['for_합계'] * df.iloc[i - 82]['for_o60'] / df.iloc[i - 82]['for_합계']
```

```python
# 나머지경찰서 2020(하반기)
for i in range(359, 368):
    df[i:i + 1]['popu_u20'] = df.iloc[i - 41]['popu_합계'] * df.iloc[i - 82]['popu_u20'] / df.iloc[i - 82]['popu_합계']
    df[i:i + 1]['popu_2030'] = df.iloc[i - 41]['popu_합계'] * df.iloc[i - 82]['popu_2030'] / df.iloc[i - 82]['popu_합계']
    df[i:i + 1]['popu_4050'] = df.iloc[i - 41]['popu_합계'] * df.iloc[i - 82]['popu_4050'] / df.iloc[i - 82]['popu_합계']
    df[i:i + 1]['popu_o60'] = df.iloc[i - 41]['popu_합계'] * df.iloc[i - 82]['popu_o60'] / df.iloc[i - 82]['popu_합계']

    df[i:i + 1]['for_u20'] = df.iloc[i - 41]['for_합계'] * df.iloc[i - 82]['for_u20'] / df.iloc[i - 82]['for_합계']
    df[i:i + 1]['for_2030'] = df.iloc[i - 41]['for_합계'] * df.iloc[i - 82]['for_2030'] / df.iloc[i - 82]['for_합계']
    df[i:i + 1]['for_4050'] = df.iloc[i - 41]['for_합계'] * df.iloc[i - 82]['for_4050'] / df.iloc[i - 82]['for_합계']
    df[i:i + 1]['for_o60'] = df.iloc[i - 41]['for_합계'] * df.iloc[i - 82]['for_o60'] / df.iloc[i - 82]['for_합계']
```

```python
df.to_csv('14번.연령별인구_total_f.csv', index = 0)
```
