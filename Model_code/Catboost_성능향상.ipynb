# Catboost
- 기존 부스팅 과정과 전체적인 양상은 비슷하다
- 기존 부스팅은 모델이 일괄적으로 모든 훈련 데이터를 대상으로 잔차를 계산한다
- Catboost는 일부만 가지고 잔차 계산을 하고 모델을 만들고, 그 뒤에 있는 데이터의 잔차를 이 모델의 예측값을 사용한다
- 예를들어 x1, x2, x3 ----, x10 이 있다고 하면, x1으로 잔차 계산 -> 모델 생성 -> x2를 예측값으로 사용,
  x1,x2으로 잔차 계산 -> 모델 생성 -> x3,x4를 예측값으로 사용 --- 반복
- 위의 과정을 진할 할 때, 데이터를 셔플링하며 뽑아낸다.(각기 다른 트리로 만든다)
- 데이터수에 따라 가지수가 많아지므로, 지금처럼 적은 데이터에 효과적이다
- has_time = True로 설정하면 시간이 지남에 따른 것도 반영 가능한다
