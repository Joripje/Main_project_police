````python
# 10대 미만은 10대로 통일, 80대 이상 = 80대 + 90대
def set_agesex(data, col):
    
    for i in range(data.shape[0]):
        if data.loc[i,col] == '10대 미만':
            data.loc[i, col] = '10대'
            
    for i in range(data.shape[0]):
        if data.loc[i, col] == '90대 이상':
            data.loc[i,col] = '80대'
            
    for i in range(data.shape[0]):
        if data.loc[i, col] == '기타불명':
            data.loc[i,col] = '미분류'
            
    return(data)
````

````python
accident = set_agesex(accident, 'ofn_age')

accident = set_agesex(accident, 'vic_age')
accident = set_agesex(accident, 'ofn_age')

accident = set_agesex(accident, 'vic_sx')
````

````python
def sum_acc_case(data):
    data['차대사람'] = data[['차대사람-기타', '차대사람-길가장자리구역통행중','차대사람-보도통행중',
                        '차대사람-차도통행중','차대사람-횡단중']].sum(axis = 1)
    
    data['차대차'] = data[['차대차-기타','차대차-정면충돌','차대차-추돌','차대차-측면충돌', '차대차-후진중충돌']].sum(axis=1)
    
    data["차량단독"] =  data[['차량단독-공작물충돌', '차량단독-기타', '차량단독-도로외이탈-기타', '차량단독-도로외이탈-추락',
       '차량단독-전도전복-전도', '차량단독-전도전복-전복', '차량단독-주/정차차량충돌']].sum(axis=1)
    
    data = data[['jur_stn','date','차대사람','차대차','차량단독']]
    
    return data
````

````python
acs_18_1 = sum_acc_case(acs_18_1)
acs_18_2 = sum_acc_case(acs_18_2)
acs_19_1 = sum_acc_case(acs_19_1)
acs_19_2 = sum_acc_case(acs_19_2)
acs_20_1 = sum_acc_case(acs_20_1)
acs_20_2 = sum_acc_case(acs_20_2)
````
