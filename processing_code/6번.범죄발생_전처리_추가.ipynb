```python
import pandas as pd
import numpy as np
df6 = pd.read_csv("./data/6번.범죄발생_total.csv")
base = pd.read_csv("./data/base_df.csv")
```

```python
df6
```

```python
base.columns = ["jur_stn", "year"]
```

```python
df6[df6["jur_stn"] == "마산동부경찰서"]
```

```python
df_mer = pd.merge(df6, base, how = "outer", on = ["jur_stn", "year"])
df_mer
```

```python
for i in list(base["jur_stn"].unique()):
    df_mean = round(df_mer.loc[(df_mer["jur_stn"] == i) & (df_mer["year"].str.contains("상반기"))].mean()).copy()
    df_mer.loc[(df_mer["jur_stn"] == i) & (df_mer["year"].str.contains("상반기"))] = df_mer.loc[(df_mer["jur_stn"] == i) & (df_mer["year"].str.contains("상반기"))].fillna(df_mean)
    
for i in list(base["jur_stn"].unique()):
    df_mean = round(df_mer.loc[(df_mer["jur_stn"] == i) & (df_mer["year"].str.contains("하반기"))].mean()).copy()
    df_mer.loc[(df_mer["jur_stn"] == i) & (df_mer["year"].str.contains("하반기"))] = df_mer.loc[(df_mer["jur_stn"] == i) & (df_mer["year"].str.contains("하반기"))].fillna(df_mean)
```
